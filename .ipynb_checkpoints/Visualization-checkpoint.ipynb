```python
import matplotlib.pyplot as plt
import pickle

tau = [0.1, 0.2, 0.3, 0.5, 0.7,0.9,1.0]
widths = [30,62,126,254,510]
keys = ['train_loss', 'train_acc@1', 'train_acc@5', 'test_loss', 'test_acc@1', 'test_acc@5']
test_acc1 = {}
test_acc5 = {}


for ss in tau:
    rootdir = f'linear_ss{ss}_results'
    test_acc1[ss] = []
    test_acc5[ss] = []
    for w in widths:
        resultdict = pickle.load(open(f'./{rootdir}/linearN{w}.p', 'rb'))
        test_acc1[ss].append(resultdict['test_acc@1'][-1])
        test_acc5[ss].append(resultdict['test_acc@5'][-1])
    
```

```python

for ss in tau:
    plt.plot(widths, test_acc1[ss], label=f'N={int(ss*100000):,}', linestyle='--', marker='o')
    
plt.xlabel('Number of negative samples (N)')
plt.ylabel('Top-1 Accuracy')
plt.title('Effects of negative sampling (STL10 on ResNet10) -- N = # of training data used ')
plt.legend(title='Legends', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.show()
```
